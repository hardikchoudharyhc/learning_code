Type of data 
1. Numerical
2. Categorical

PLotting a graph on-
a. One column is univariate analysis
b. Two column is Bi-variate analysis
c. More than two columns is Multi-variate analysis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

Import libraries.

### 2D Line plot

Bivariate Analysis
categorical -> numerical and numerical -> numerical
Use case - Time series data

```python
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Price of Bitcoin over the years')
plt.show()
plt.xlim(2016,2021)
plt.ylim(40000,60000)
plt.plot(year,price)
plt.show() 
```

```python
batsmen = pd.read_csv('sharma-kohli.csv')
plt.plot(batsmen['index'],batsmen['V Kohli'], color='red', label='V Kohli', marker='o', markersize=5)
plt.plot(batsmen['index'],batsmen['RG Sharma'], color='blue', label='RG Sharma', linestyle='dashdot', linewidth=3)
plt.xlabel('Years')
plt.ylabel('Runs')
plt.title('Runs scored by RG Sharma and V Kohli over the years')
plt.legend(loc='best')
plt.show()
```

Scatterplots. Uses: -

Bar chart
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups



```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']
plt.bar(colors,children,color='black')
```

```python

# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')
# Remove existing labels and add numerical labels and perform shift to the left by 0.2

plt.xticks(np.arange(df.shape[0]), df['batsman'])
# plt.xticks(rotation=90(or vertical))
# Change the x-axis labels to the batsman names

plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.title('Runs scored by top 10 batsman in 2015, 2016 and 2017')
plt.legend(['2015','2016','2017'])
plt.show()
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

Histogram

- Use case - Frequency Count

```python
# simple data

data = [32,45,56,10,15,27,61]

# plt.hist(data)
# Can provide bins
plt.hist(data,bins=[10,20,30,40,50,60,70])
```

Piechart
- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```
