Importing required libraries.

```python
import pandas as pd
import requests
```

Getting the response of API call using requests.get() and storing it in response variable.

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

Creating a temporary dataset variable and storing the required data of response.

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
temp_df
```

Visiting all the other pages and collecting the data from their via a for loop, and is stored as a list in dfs, these list are converted into a dataset using pd.concat which concat each list in dfs vertically making them a row.

```python

dfs = []

for i in range(1, 429):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}')

    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]

    dfs.append(temp_df)

df = pd.concat(dfs, ignore_index=True)
df
```

The dataframe is converted in .csv format and taken as output file.

```python
df.to_csv('movies.csv')
```
